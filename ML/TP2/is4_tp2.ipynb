Intéressons nous aux paramètres : max_depth, max_features and random_state.

Fixons ```random_state=2``` et  ```max_features=None``` puis faisons prendre les valeurs {1,2,4,None} à ```max_depth```.


Considérons deux ensembles _train_ et _test_ comme construits précédement à partir de la totalité du jeux de données (```random_state=3```) en faisant varier les valeurs de ```test_size``` dans l'ensemble {0.20,0.33,0.5}.

**§ Reporter dans un tableau les résultats d'accuracy pour ces valeurs de paramètres pour les deux jeux de données : iris puis digits**

**§ Que remarquez-vous ? Donner votre analyse des résultats**

````python
#Il n'est pas nécessaire de normaliser les données car pas besoin dans les arbres de décisions
#arbre de décision prend décision sur les colonnes
#svm prend en compte des pondérations->
#methodes apprentissage algébrique->privilégier normalisation

from sklearn import tree
from sklearn.tree import export_text
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np

#random_stateControls the randomness of the estimator. The features are always randomly permuted at each split,
#even if ``splitter`` is set to ``"best"``. When ``max_features < n_features``, the algorithm will select 
# ``max_features`` at random at each split before finding the best split among them.

#max_features : The number of features to consider when looking for the best split

#max_depth:The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or 
#until all leaves contain less than min_samples_split samples.

#help(tree.DecisionTreeClassifier)

from sklearn.datasets import load_iris
iris = load_iris()
Xiris, yiris = iris.data, iris.target
test=[0.2,0.33,0.5]
tab=np.zeros((3,4))
indx=0
for i in test:
    X_trainiris, X_testiris, y_trainiris, y_testiris = train_test_split(Xiris, yiris, test_size=i, random_state=3)
    dep=[1,2,4,None]
    indy=0
    for j in dep:
        clf=tree.DecisionTreeClassifier(random_state=2, max_features=None,max_depth=j)
        clf=clf.fit(X_trainiris,y_trainiris)
        y_pred = clf.predict(X_testiris)
        tab[indx,indy]=accuracy_score(y_testiris,y_pred) * 100.0
        indy=indy+1
    indx=indx+1

np.set_printoptions(precision=2, suppress=True)

# Display tab with row and column indices
print("   ", end="")
for j in dep:
    print(f"depth={j}", end="\t")
print()

for i in range(tab.shape[0]):
    print(f"test_size={test[i]}", end="\t")
    for j in range(tab.shape[1]):
        print(f"{tab[i, j]:.2f}", end="\t")
    print()

````

````python
from sklearn.datasets import load_digits
digits = load_digits()
X, y = digits.data, digits.target
test=[0.2,0.33,0.5]
tab=np.zeros((3,4))
indx=0
for i in test:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=i, random_state=3)
    dep=[1,2,4,None]
    indy=0
    for j in dep:
        clf=tree.DecisionTreeClassifier(random_state=2, max_features=None,max_depth=j)
        clf=clf.fit(X_train,y_train)
        y_pred = clf.predict(X_test)
        tab[indx,indy]=accuracy_score(y_test,y_pred) * 100.0
        indy=indy+1
    indx=indx+1

np.set_printoptions(precision=2, suppress=True)

# Display tab with row and column indices
print("      ", end="")
for j in dep:
    print(f"depth={j}", end="\t")
print()

for i in range(tab.shape[0]):
    print(f"test_size={test[i]}", end="\t")
    for j in range(tab.shape[1]):
        print(f"{tab[i, j]:.2f}", end="\t")
    print()
````

Maintenant intéressons nous à la méthodologie d'entrainement avec un jeu de données. Jusqu'ici nous avons utilisé la méthode qui consiste à séparer le jeu de données pour obtenir un jeu d'entrainement et un jeu de test. Etudions ici l'intérêt de la [validation croisée](https://scikit-learn.org/stable/modules/cross_validation.html#cross-validation). Pour cela, il est nécessaire d'importer un module via la commande 
```from sklearn.model_selection import cross_val_score```.

Pour commencer, instancions un arbre de décision comme suit : 
```dt = tree.DecisionTreeClassifier(random_state=2,max_depth=4)```

Pour obtenir (et afficher) le résultat de cette instanciation sur le jeu de données (complet) X et la classe associée y par validation croisée, il suffit d'exécuter la commande : 
```cross_val_score(dt, X, y, cv=5)```

**§ Regarder la doc en ligne du module ```cross_val_score``` et expliquer la méthode de validation croisée effectuée**

**§ Comparer ce résultat à celui obtenu précédement pour la même version du classifier**

````python
from sklearn.model_selection import cross_val_score
dt = tree.DecisionTreeClassifier(random_state=2,max_depth=4)
cross_val_score(dt, X, y, cv=5)
#cv détermine la stratégie de cross-validation
#cv=5 spécifie le nombre de fold
#help(cross_val_score)

#Pour comparer à la même version du classifier, il faut comparer à la colonne depth=4
#on remarque qu'on est meilleur avec le classifier
````

    valide sur 1 observation pour chaque fold
    bien si petit jeu de donnée pour s'entrainer, sinon pas terrible
    

Etudions aussi d'autres modèles classiques de validation croisée :
* K-fold (commande ```from sklearn.model_selection import KFold```)
* Repeated K-fold (commande ```from sklearn.model_selection import RepeatedKFold```)
* Repeated Stratified K-fold (commande ```from sklearn.model_selection import RepeatedStratifiedKFold```)
* LeaveOneOut (commande ```from sklearn.model_selection import LeaveOneOut```)

**§ Regarder la doc en ligne et expliquer ces quatre méthodes de validation croisée**

````python
from sklearn.model_selection import KFold
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import LeaveOneOut
````

1. KFold : En validation croisée K-fold, le jeu de données est divisé en K sous-ensembles (plis), et le modèle est entraîné K fois, chaque fois en utilisant K-1 plis pour l'entraînement et le pli restant pour la validation. Le processus est répété jusqu'à ce que chaque pli ait été utilisé comme données de validation exactement une fois. 

2. RepeatedKFold : La validation croisée K-fold répétée étend la K-fold en répétant le processus plusieurs fois avec différentes partitions aléatoires des données. Cela permet de réduire la variance des métriques d'évaluation.

3. RepeatedStratifiedKFold : Cette méthode est similaire à la K-fold répétée, elle garantit que la distribution des classes dans chaque pli est similaire à la distribution globale des classes. C'est particulièrement utile pour les ensembles de données déséquilibrés.

4. LeaveOneOut : chaque observation est utilisée comme ensemble de validation exactement une fois. C'est équivalent à la validation croisée K-fold avec K égal au nombre d'échantillons dans le jeu de données.


Tester ces méthodes avec l'arbre de décision précédent et les valeurs ```n_splits=5``` et ```n_repeats=5``` lorsque c'est applicable. 
Par exemple pour la méthode K-fold, la commande ```kf = KFold(n_splits=5)``` instancie un objet kf afin de faire une validation croisée 5-fold et la commande ```cross_val_score(dt, X, y, cv=kf)``` exécute la validation croisée 5-fold.
La moyenne des accuracy est accessible via la méthode ```.mean()``` du module ```cross_val_score```.

**§ Exécuter les 5 méthodes et analyser les résultats**

**§ Quel est l'intérêt de la validation croisée par rapport au découpage train/test du jeu de données ?**

````python
#methode classique
print(cross_val_score(dt, X, y, cv=5).mean())

#KFold
kf = KFold(n_splits=5)
print(cross_val_score(dt, X, y, cv=kf).mean())

#RepeatedKFold
kf2=RepeatedKFold(n_repeats=5)
print(cross_val_score(dt, X, y, cv=kf2).mean())

#RepeatedStratifiedKFold
kf3=RepeatedStratifiedKFold(n_repeats=5,n_splits=5)
print(cross_val_score(dt, X, y, cv=kf3).mean())

#LeaveOneOut
kf4=LeaveOneOut()
print(cross_val_score(dt, X, y, cv=kf4).mean())
````

La validation croisée va nous permettre d'utiliser l'intégralité de notre jeu de données pour l'entraînement et pour la validation !
