```python
# Import des bibliothèques
from sklearn.model_selection import cross_val_score, RepeatedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_digits
import numpy as np
import time

# Chargement de la base de données Iris
digits = load_digits()
X = digits.data
y = digits.target

# Définition des valeurs à tester
n_estimators_values = [50, 100, 200]
max_depth_values = [1, 2, 4, None]

num_repeats = 30
num_splits = 5  # Number of folds in k-fold cross-validation

tab = np.zeros((len(n_estimators_values) * len(max_depth_values), 4))
row_index = 0

for n_estimators in n_estimators_values:
    for max_depth in max_depth_values:
        start = time.time()

        # Create RandomForestClassifier with specified parameters
        rf_model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth)

        # Create RepeatedKFold with specified number of repeats and splits
        rkf = RepeatedKFold(n_repeats=5)

        # Perform cross-validation
        accuracy_scores = cross_val_score(rf_model, X, y, cv=rkf)
        average_accuracy = np.mean(accuracy_scores)

        end = time.time()
        average_execution_time = (end - start) / num_repeats

        # Store results in the tab array
        tab[row_index, 0] = n_estimators
        tab[row_index, 1] = max_depth
        tab[row_index, 2] = average_accuracy * 100.0
        tab[row_index, 3] = average_execution_time

        row_index += 1

# Display tab with row and column indices
print("n_estimators\tDepth\t\tAvg Accuracy\tAvg Time")
for i in range(tab.shape[0]):
    print(f"{int(tab[i, 0])}\t\t{tab[i, 1]}\t\t{tab[i, 2]:.2f}%\t\t{tab[i, 3]:.4f} seconds")


```
